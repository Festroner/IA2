```python
import random
import string
import math
import time

random.seed=random.seed(int(round(time.time())))


Frase='Me ll'.lower()
numPoblacion=5
generaciones=10
pMutacion=0.9
resumen=2
muestra=5

#Genes posibles
letras=string.ascii_lowercase + ' '
```

```python

def coincidencias(cadena,original):
    num=0
    for i in range(len(original)):
        if cadena[i]==original[i]:
            num+=1
    return num


def CalculoFitness(cadena, original, coincidencias):
    return math.exp(coincidencias - len(original)) - math.exp(-len(original))


def Ruleta(lista, coincide):
    #la funcion recibe un array con los fitness de toda la poblacion y otro con las coincidencias de cada cromosoma

#LA RULETA CREARÁ UN ARRAY DE 100 POSICIONES. CADA POSICION CONTENDRÁ EL NUMERO DEL ELEMENTO QUE CORRESPONDE
#HABRÁ MÁS NUMEROS DE UN ELEMENTO DEPENDIENDO DE SU FITNESS
#LA FUNCION DEVUELVE EL NUMERO DEL GEN ELEGIDO AL AZAR

    prob=[x/(math.exp(max(coincide)-len(Frase))-math.exp(len(Frase))) for x in lista] #probabilidad de selecion
    #print(prob)
    porcentajes=[x/sum(prob) for x in prob] #calculo de porcentajes
    #print(porcentajes)
    array=[]
    
    for i in range(len(porcentajes)):
        array.append(MonteCarloSimple([1-porcentajes[i],porcentajes[i]]))


    #for i in range(len(adapta)): #Para el numero del elemento
    #    for j in range(adapta[i]):#numero de veces que se escribe
    #        array.append(i)
    #while len(array)<100:
    #    array.append(random.randrange(len(lista)))
    #
    #elegido=random.randrange(99)


    print("array",len(array))
    print(array)
    return array

def MonteCarloSimple(lista):#Creamos la ruleta por sectores y elegimos un numero al azar max 100
    ruleta=[]
    lista=[int(round(x*100)) for x in lista]
    for i in range(len(lista)):
        for j in range(lista[i]):
            ruleta.append(i)
    #print(max(ruleta))
    elegido=random.randrange(len(ruleta))
    return ruleta[elegido] # Esta funcion devuelve un 1 si solo son 2 valores (1 positivo, 0 negativo)

def Mutacion(padre):#a traves de la frase del padre evaluamos uno a uno si los caracteres cambian
    nueva=padre
    for j in range(len(Frase)):
        if MonteCarloSimple([(1-pMutacion),pMutacion]) == 0:
            sustituto=random.choice(letras)
            #print('Letra al azar:    ',sustituto)
            nueva=nueva.replace(nueva[j],sustituto)
                    
    return nueva



```

```python
primera=True
epocas=0
poblacion=[]
PoblacionAnterior=[]
probabilidadP=[]
probabilidadPAnterior=[]
fitnesAnterior=[]
coincideAnterior=[]

while True:
    fitnes=[]
    coincide=[]
    #print(epocas)
 
    while primera and epocas == 0:
        poblacion=[]
        coincide=[]
        #Poblacion aleatoria
        for i in range(numPoblacion):
            frase=''
            
            for j in range(len(Frase)):
                frase+=random.choice(letras)
            poblacion.append(frase)
            coincide.append(coincidencias(poblacion[i], Frase))
        if max(coincide)!=0:
            primera=False

    #Calculamos el fitness
    for i in range(numPoblacion):
        if epocas!=0:
            coincide.append(coincidencias(poblacion[i], Frase))#contamos el numero de coincidencias
        resul=CalculoFitness(poblacion[i], Frase, coincide[i])#calculamos el fitnes de cada individuo
        fitnes.append(resul)


    

    #guardamos el fitnes 
    #print(fitnes)

    if max(fitnes) == 0:#si el maximo fitnes es = 0 recuperamos la generacion anterior y la mutamos
        poblacion=PoblacionAnterior
        probabilidadP=probabilidadPAnterior
        fitnes=fitnesAnterior
        coincide=coincideAnterior
    else:

        probabilidadP=[x/sum(fitnes) for x in fitnes]
    
    #guardamos la generacion
    coincideAnterior=coincide
    PoblacionAnterior=poblacion
    probabilidadPAnterior=probabilidadP
    fitnesAnterior=fitnes 

    #Seleccion del padre
    Padre=MonteCarloSimple(probabilidadP)

    #probabilidad de seleccion para mutacion para cada cromosoma
    PSeleccion=Ruleta(fitnes,coincide)
    #hijo al azar por el que empezamos a ver quien muta
    hijo1=random.randrange(numPoblacion-1)




    for i in range(hijo1,len(poblacion)):
        if PSeleccion[i] == 1:
            poblacion[i]=Mutacion(poblacion[Padre])
    for i in range(hijo1):
        if PSeleccion[i] == 1:
            poblacion[i]=Mutacion(poblacion[Padre])
    #Elegimos al hijo y decidimos si sus genes mutan o no
    

    if epocas%5 == 0:
        print("\n\nGeneraciones:",epocas, poblacion)
        print(coincide)

    if epocas == generaciones or max(coincide)==len(Frase):
        break
    epocas+=1
    
#Ruleta(fitnes)


```
