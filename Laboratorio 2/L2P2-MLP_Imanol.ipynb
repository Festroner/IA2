```python
import pandas as pd
import numpy as np

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras import optimizers


#FUNCION DE LA PUERTA XOR
def XOR (a, b):
    if a != b:
        return 1
    else:
        return 0

#DATOS PARA ENTRENAR
salidas=[]
entradas= np.random.randint(0, 2, (200, 2))
for i in range(len(entradas)):
    salidas.append(XOR(entradas[i][0],entradas[i][1]))
    
salidas=np.array(salidas)


#DATOS PARA VALIDAR
evalua=[]
entrenamiento= np.random.randint(0, 2, (50, 2))
for i in range(len(entrenamiento)):
    evalua.append(XOR(entrenamiento[i][0],entrenamiento[i][1]))
    
evalua=np.array(evalua)

```

```python
#CREAMOS EL MODELO
model= Sequential()
#AÑADIMOS UNA CAPA OCULTA CON 3 NEURONAS CONECTADA CON LAS 2 ENTRADAS
#COMO LA SOLUCION ES NO LINEAL UTILIZAMOS LA FUNCION SIGMOIDEA
model.add(Dense(3, 'sigmoid', 2))
model.add(Dense(1, 'sigmoid'))# UNA SALIDA (0/1)

#ERROR MEDIO CUADRÁTICO, Adam, metrica: calcula las veces que el modelo acierta
model.compile(loss="mean_squared_error", optimizer='adam',metrics="accuracy")
model.fit(x=entradas,y=salidas,validation_split= 0.1, epochs=1000, shuffle=False)
```

```python
scores = model.evaluate(entrenamiento, evalua)
print("\n%s: %.2f%%" % (model.metrics_names[1], scores[1]*100))
```

```python

#comprobamos la efectividad del modelo con datos nuevos
prediccion=np.random.randint(0, 2, (100, 2))
esperado=[]
for i in range(len(prediccion)):
    esperado.append(XOR(prediccion[i][0],prediccion[i][1]))
esperado=np.array(esperado)
obtenido=model.predict(prediccion).round()

aciertos=0

for i in range(len(esperado)):
    if obtenido[i]==esperado[i]:
        aciertos+=1

print("El modelo ha acertado ", aciertos, "de 100 casos propuestos")
```
